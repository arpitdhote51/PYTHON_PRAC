```python
# prompt: Line Plot: Stock Prices Over Time ðŸ“Œ Problem: Plot a line graph showing stock prices for a company over 10 days.
# X-axis: Days (1 to 10) Y-axis: Stock Prices (e.g., [100, 102, 105, 103, 107, 110, 108, 115, 120, 118]) Customize the graph with labels, title, and a grid.

import matplotlib.pyplot as plt

days = list(range(1, 11))
stock_prices = [100, 102, 105, 103, 107, 110, 108, 115, 120, 118]

plt.plot(days, stock_prices)

plt.xlabel("Days")
plt.ylabel("Stock Prices")
plt.title("Stock Prices Over 10 Days")
plt.grid(True)

plt.show()
```
